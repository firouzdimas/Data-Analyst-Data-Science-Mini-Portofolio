```python
labels = df['Furnishing Status'].value_counts().index
values = df['Furnishing Status'].value_counts()

fig = go.Figure(data=[go.Pie(labels=labels, values=values)])
fig.show()
```

```python
labels = df['Tenant Preferred'].value_counts().index
values = df['Tenant Preferred'].value_counts()

fig = go.Figure(data=[go.Pie(labels=labels, values=values)])
fig.show()
```
